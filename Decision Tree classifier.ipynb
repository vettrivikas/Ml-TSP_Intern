## ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iri = pd.read_csv('Iris.csv')
iri.head()
```

```python

pd.plotting.parallel_coordinates(iri.iloc[:,1:],"Species", colormap=plt.get_cmap("Set2"))
plt.show
```

```python
iris = iri.drop(['Id','Species'],axis=1)
iris.head()
```

```python
iris.info()
```

```python
iri['Species'].value_counts()
```

```python

from sklearn.preprocessing import LabelEncoder
enco = LabelEncoder()
target = enco.fit_transform(iri['Species'])
target
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(iris,target)

print("Decision Tree Classifier created")
```

```python
fl = [x for x in iris.columns]
ll = ['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(20,25))
_ = tree.plot_tree(clf, 
                   feature_names= fl,  
                   class_names= ll,
                   filled=True)
```
